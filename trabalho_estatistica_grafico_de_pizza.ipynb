```r
# Região Sudeste

valores_adicionados_Sudeste = filter(valores_adicionados, Nome.da.Grande.Região=="Sudeste")

head(valores_adicionados_Sudeste)

valores_agropecuaria_Sudeste = sum(valores_adicionados_Sudeste$Valor.adicionado.bruto.da.Agropecuária...a.preços.correntes..R..1.000.)
valores_agropecuaria_Sudeste

print("")

valores_industria_Sudeste = sum(valores_adicionados_Sudeste$Valor.adicionado.bruto.da.Indústria..a.preços.correntes..R..1.000.)
valores_industria_Sudeste

print("")

valores_servicos_Sudeste = sum(valores_adicionados_Sudeste$Valor.adicionado.bruto.dos.Serviços..a.preços.correntes....exceto.Administração..defesa..educação.e.saúde.públicas.e.seguridade.social..R..1.000.)
valores_servicos_Sudeste

print("")

valores_administracao_Sudeste = sum(valores_adicionados_Sudeste$Valor.adicionado.bruto.da.Administração..defesa..educação.e.saúde.públicas.e.seguridade.social...a.preços.correntes..R..1.000.)
valores_administracao_Sudeste
```

```r
dados <- data.frame(região, agropecuaria, industria, servicos, administracao)

# Selecionar apenas as colunas relevantes para o gráfico de pizza
dados_pizza <- dados[, c("agropecuaria", "industria", "servicos", "administracao")]

# Calcular o total por região
dados_pizza$total <- rowSums(dados_pizza[, -1])

# Configurar as cores para cada setor da pizza
cores <- c("#FF8080", "#80FF80", "#8080FF", "#FFFF80")

# Criar o gráfico de pizza
pie(dados_pizza$total, labels = dados$região, col = cores, main = "Distribuição por Região")

# Adicionar legenda
legend("topright", legend = names(dados_pizza)[-5], fill = cores, cex = 0.8)
```

```r
# Instale o pacote ggplot2 se ainda não o tiver instalado
# install.packages("ggplot2")

library(ggplot2)

# Crie os vetores de categoria e valor corretamente
categorias <- c("Agropecuaria", "Industria", "Servicos", "Administracao")
valores <- c(dados_Norte$agropecuaria, dados_Norte$industria, dados_Norte$servicos, dados_Norte$administracao)

# Crie o dataframe com os dados
dados_Norte <- data.frame(
  Categoria = categorias,
  Valor = valores
)



# Crie o gráfico de pizza
grafico <- ggplot(dados_Norte, aes(x = "", y = Valor, fill = Categoria)) +
  geom_bar(stat = "identity", width = 1) +
  coord_polar(theta = "y") +
  labs(fill = "Categoria") +
  theme_void() +
  geom_text(aes(label = Valor), position = position_stack(vjust = 0.5), color = "white", size = 6)

# Exiba o gráfico
print(grafico)
```
