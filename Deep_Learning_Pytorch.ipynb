# Tipos de tensores:

Você pode criar tensores do PyTorch de inúmeras formas! Vamos ver primeiro os tipos de tensores que estão ao nosso dispor. Para isso, vamos converter listas comuns do Python em tensors do PyTorch.

```python
# importando o framework
import torch

lista = [[1,2,3],
         [4,5,6]]

# transformando uma lista em um tensor do pytorch

tsn = torch.Tensor(lista)
print(tsn)
print(tsn.dtype)

# Percebe-se que o torch transforma a lista
# num tensor, sendo que por padrão os valores são
# transformados em tipo float, como podemos perceber
# quando printamos o datatype.
```

```python
# Podemos ainda forçar o pytorch a transformar 
# os valores em float ou inteiros

import torch

lista = lista
tsn = torch.DoubleTensor(lista)

print(tsn)

# Para valores em float.
```

```python
import torch

lista = lista
tsn = torch.LongTensor(lista)

print(tsn)

# Para valores em inteiros. 
```

# Tensores inicializados. 

Essas funções recebem como parâmetro o tamanho de cada dimensão do tensor. Aqui vamos conhecer as seguintes funções:

torch.ones() -> Cria um tensor preenchido com uns.

torch.zeros() -> Cria um tensor preenchido com zeros.

torch.randn() -> Cria um tensor preenchido com números aleatórios a partir de uma distribuição normal.

```python
# Como vemos, através dessas instancias, criamos
# tensores passando como parâmetros a quantidade de linhas e colunas, 
# como o primeiro parâmetro e o segundo, respectivamente. 

tns1 = torch.ones(2,3)
tns0 = torch.zeros(4,5)
tnsr = torch.randn(2,5)

print(tns1)
print("")

print(tns0)
print("")

print(tnsr)
```

# Operações com tensores:

A função .item() utilizada anteriormente extrai o número de um tensor que possui um único valor, permitindo realizar as operações numéricas do Python. Caso o item não seja extraído, operações que envolvam tensores vão retornar novos tensores.

Vale ressaltar também que operações entre tensores são realizadas ponto a ponto, operando cada elemento (i, j) do tensor t1, com o elemento (i, j) do tensor t2, por isso retornando um novo tensor.

```python
# Quando eu gero um tensor com um valor além do da linha e da coluna, 
# que aqui é valor entre o primeiro 2 e o 3, esse corresponde ao valor
# de profundidade, que significa que será gerada mais um tensor, como 
# mostrado nos prints abaixo. 

tns1 = torch.randn(2,2,3)
tns2 = torch.ones(2,2,3)

print(tns1)

print("")

print(tns2)

print("")

print(tns1*tns2)
```

```python
# A função size se assemelha au shape() do numpy; 
print(tns2.size())
```

```python
# A função view no pytorch tem a função de 
# transformar o tensor, podendo achatá-lo através 
# do -1, que é isso que esse valor simboliza. 
print(tns2.view(tns2.size(0), -1))
```

```python
# Além de achatar eu posso redimensionar o 
# array anterior num novo, passando como parâmetro, por exemplo, 
# a quantidade de linhas e de colunas. 
print(tns2.view(4,3))
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = -2
b = 5 
c = 5

# ax + by + c = 0 
# (-ax - c)/b = y

def equacao_da_reta(a, b, c):

  x = np.linspace(0,10, 10)
  y = (-a*x -c)/b

  # Constroi um eixo vertical e horizontal respectivamente
  plt.axvline(0, -1, 4, color='k', linewidth=1)
  plt.axhline(0, -2, 10, color='k', linewidth=1)
  plt.plot(x, y)
  plt.grid(True)

equacao_da_reta(a,b,c)

from matplotlib import markers
# Visualizando diferentes pontos no gráfico:

p1 = (2,0.4)
p2 = (5, 1)
p3 = (3,-0.5)

plt.plot(p1[0], p1[1], color="b", marker="o")
plt.plot(p2[0], p2[1], color="g", marker="o")
plt.plot(p3[0], p3[1], color="r", marker="o")
```

```python
reta1 = a*p1[0] + b*p1[1] + c
print("O resultado da função é igual a: %.2f" % reta1)
print("")
reta2 = a*p2[0] + b*p2[1] + c
print("O resultado da função é igual a: %.2f" % reta2)
print("")
reta3 = a*p3[0] + b*p3[1] + c
print("O resultado da função é igual a: %.2f" % reta3)
```

```python
# Aqui podemos ver o resultado de cada função, sendo que para os valores 
# positivos sabemos que corresponde a um elemento positivo, uma vez 
# que se encontra acima da reta, enquanto que os valores iguais a zero indicam
# que estão sob a reta e, por fim, os negativos indicam que estão abaixo da reta. 

# A classificação linear é boa para quando temos dados que podem ser separados
# através da reta da função. Por exemplo, separando gato de cachorro, o algoritimo
# iria criar, na classificação linear, uma função da reta, a qual separaria os
# valores gato ou cachorro, divindo-os a partir de si. Assim o gato, por exemplo, 
# poderia estar acima e o cachorro em baixo e vice-versa, como pode ser visto abaixo. 
```

```python
SEED = 8
np.random.seed(SEED)
plt.scatter(x[:,0], x[:,1], marker="o", c=y, edgecolors="k")

def modelo_linear(w1, w2, b):

  # Aqui estou traçando um limite
  # para a visualização do gráfico 
  # nos eixos x e y. 
  xmin, xmax = plt.gca().get_xlim()
  ymin, ymax = plt.gca().get_ylim()

  x = np.linspace(-3,3, 10)
  y = (-w1*x -b)/w2

  # Constroi um eixo vertical e horizontal respectivamente
  plt.axvline(0, -1, 4, color='k', linewidth=1)
  plt.axhline(0, -2, 10, color='k', linewidth=1)
  plt.plot(x, y)
  plt.grid(True)
 
  # Aqui estou plotando as delimitações dos eixos. 
  plt.xlim(xmin,xmax)
  plt.ylim(ymin,ymax)

w1 = 8
w2 = 2 
b = 0.5

modelo_linear(w1,w2,b)
```

# Visualizando a acurácia desse modelo linear:



```python
# Criando uma função que classifica os pontos que a ele
# são passados e depois retornando, para verificar, 
# seu rótulo e cor. 

def classificacao(ponto_cartesiano, w1, w2, b):

  reta = ponto_cartesiano[0]*w1 + ponto_cartesiano[1]*w2 + b

  if reta >=0:
    return 1, "yellom"
  else:
    return 0, "purple"

ponto_cartesiano = (-2,2)

classificacao(ponto_cartesiano, w1,w2,b)
```

```python
# Para eu visualizar a acurácia do modelo, devo passar cada item dos dados,
# que são acessados por meio da variável x, a qual fornece a coordenada em x e em y, 
# na função classificação, que como seu x terá o valor de cada k dentro do ciclo for. 
# Caso a categoria, que significa o rótulo, seja igual ao rótulo de cada k, haverá um 
# incremento de acertos os quais ao final são submetidos a razão de todos os dados
# para termos a acurácia. 

acertos = 0
for k in range(len(x)):
  categoria, _ = classificacao(x[k], w1, w2, b)
  if categoria == y[k]:
    acertos = acertos + 1

acuracia = (acertos/len(x))*100

print("A acurácia do modelo é de: %.2f" % acuracia)
```

Instanciando o perceptrom:

```python
import torch
from torch import nn

torch.manual_seed(50)

# instanciando o perceptron de função linear que requere entrada e saida.
perceptron = nn.Linear(in_features=3, out_features=1, bias=True)
```

```python
# Obtendo os pesos utilizados por esse perceptron, bem como 
# seu bias por meio do ciclo for. 

for nome, tensor in perceptron.named_parameters():
  print(nome, tensor)
```

```python
# Obtendo os pesos e viés do perceptron

print(perceptron.weight.data)
print(perceptron.bias.data)
```

```python
# note que temos como retorno os tensores, que são
# os valores dos pesos ou viés num conjunto de dados. 
```

```python
# Extraindo os pesos e viés e os transformando em valores
# numpy para que possam usados na expressão matemática. 
# Em relação aos pesos há um colchete, uma vez que queremos 
# acessar apenas o conjunto que contém os valores e, assim, 
# usamos o primeiro conjunto que os contém. 
w1, w2, w3 = perceptron.weight.data.numpy()[0]
b = perceptron.bias.data.numpy()

# w1*x1 + w2*x2 + w3*x3 + b = 0
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot3d(perceptron):
  
  w1, w2, w3 = perceptron.weight.data.numpy()[0]
  b = perceptron.bias.data.numpy()
  
  X1 = np.linspace(-1,1,10)
  X2 = np.linspace(-1,1,10)

  X1, X2 = np.meshgrid(X1,X2)

  X3 = (b - w1*X1 - w2*X2) / w3
  
  fig = plt.figure(figsize=(10, 8))

  ax = fig.gca(projection='3d')

  # Define a rotação no eixo z
  ax.view_init(azim=180)
  
  ax.plot_surface(X1, X2, X3, cmap='plasma')
  
```

```python
plot3d(perceptron)
```

```python
x = torch.Tensor([0, -1, 2])
y = perceptron(x)

plot3d(perceptron)
plt.plot([x[0]], [x[1]], [x[2]], color="r", marker="^")

print(y)
```

# Funções de Ativações:

```python
# Para ver as diferenças das funções de ativações irei
# usar o mesmo modelo linear de outrora;

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification

SEED = 8
np.random.seed(SEED)

X, Y = make_classification(n_features=2, n_redundant=0, n_informative=1, 
                           n_clusters_per_class=1)

def plotmodel(w1, w2, b):

  # Perceba que aqui estou usando o scatter justamento para plotar
  # o gráfico dos elementos gerados pelo make_classification. 
  plt.scatter(X[:,0], X[:,1], marker="o", c=Y, edgecolors="k")

  # Aqui estou traçando um limite
  # para a visualização do gráfico 
  # nos eixos x e y. 
  xmin, xmax = plt.gca().get_xlim()
  ymin, ymax = plt.gca().get_ylim()

  x = np.linspace(-1,3, 10)
  y = (-w1*x -b)/w2

  # Constroi um eixo vertical e horizontal respectivamente
  plt.axvline(0, -1, 4, color='k', linewidth=1)
  plt.axhline(0, -1, 3, color='k', linewidth=1)
  plt.plot(x, y)
  plt.grid(True)
 
  # Aqui estou plotando as delimitações dos eixos. 
  plt.xlim(xmin,xmax)
  plt.ylim(ymin,ymax)

w1 = 8
w2 = 2 
b = 0.5

plotmodel(w1,w2,b)
```

```python
# Agora irei passar as funções de ativação; 

# Aqui estou iniciando o perceptron, assim como instanciando 
# a função de ativação, que será a sigmoide
perceptron = nn.Linear(2, 1)
ativando = nn.Sigmoid()

# Aqui estou modificando os pesos e viés que seriam gerados 
# de maneira aleatória pelos que eu já havia pré-definido anteriormente. 
perceptron.weight = nn.Parameter(torch.Tensor([[w1,w2]]))
perceptron.bias = nn.Parameter(torch.Tensor([b]))

# Exibindo por meio do print os pesos e viés atribuido. 
print(perceptron.weight.data)
print(perceptron.bias.data)
```

```python
makers = ['^', 'v', '>', '<']
colors = ['g', 'b', 'r', 'y']

plt.figure(figsize=(10,8))
plotmodel(w1,w2,b)

for k, idx in enumerate([17,21,43,66]):
  
  # Aqui estarei armazenando as coordenadas de
  # cada elemento, do 17 ao 66.
  x = torch.Tensor(X[idx])

  # O que estou nomeando como retorno são os dados
  # que passaram pelo perceptron, que são retornados
  # cada quais com seu respectivo valor e, assim, 
  # passando-os para a função de ativação. 
  retorno = perceptron(x)
  funcao_ativada = ativando(retorno)

  ativacao_limiar = 0 if retorno.data < 0 else 1

  label = "retorno: {:5.2f}".format(retorno.data.numpy()[0]) + "; limiar {:4.2f}".format(ativacao_limiar) + "; função sigmoide: {:5.2f}".format(funcao_ativada.data.numpy()[0])
  plt.plot(x[0], x[1], marker = markers[k], color = colors[k], markersize=10, label = label)

plt.legend()
plt.show

# Olhando para o gráfico abaixo eu cosigo observar em como as
# funções de saida trabalham sobre a lógica da função limiar e 
# função sigmoide. Percebe-se como os valores retornados apresentam
# para cada função de saida um retorno diferente. 
```

# Observando as diferentes saidas para outros ativadores:

```python
# Aqui estou iniciando o perceptron, assim como instanciando 
# a função de ativação, que será a sigmoide
perceptron = nn.Linear(2, 1)
ativando = nn.Tanh()

# Aqui estou modificando os pesos e viés que seriam gerados 
# de maneira aleatória pelos que eu já havia pré-definido anteriormente. 
perceptron.weight = nn.Parameter(torch.Tensor([[w1,w2]]))
perceptron.bias = nn.Parameter(torch.Tensor([b]))


makers = ['^', 'v', '>', '<']
colors = ['g', 'b', 'r', 'y']

plt.figure(figsize=(10,8))
plotmodel(w1,w2,b)

for k, idx in enumerate([17,21,43,66]):
  
  # Aqui estarei armazenando as coordenadas de
  # cada elemento, do 17 ao 66.
  x = torch.Tensor(X[idx])

  # O que estou nomeando como retorno são os dados
  # que passaram pelo perceptron, que são retornados
  # cada quais com seu respectivo valor e, assim, 
  # passando-os para a função de ativação. 
  retorno = perceptron(x)
  funcao_ativada = ativando(retorno)

  ativacao_limiar = 0 if retorno.data < 0 else 1

  label = "retorno: {:5.2f}".format(retorno.data.numpy()[0]) + "; limiar {:4.2f}".format(ativacao_limiar) + "; função tangente: {:5.2f}".format(funcao_ativada.data.numpy()[0])
  plt.plot(x[0], x[1], marker = markers[k], color = colors[k], markersize=10, label = label)

plt.legend()
plt.show

# Usando o ativador tangente como função de saida, vemos que ele 
# opera segundo um intervalo de 1 e -1, assim atribuindo aos elementos 
# inscritos nesse intervalo grau de proximidade ou não. Por outro lado, 
# se eu tivesse um retorno de zero na função tangente, estaria sob um contexto 
# de incerteza, sem saber ao certo a qual classe, amarelo ou roxo, o item pertenceria. 
```

```python
# Aqui estou iniciando o perceptron, assim como instanciando 
# a função de ativação, que será a sigmoide
perceptron = nn.Linear(2, 1)
ativando = nn.ReLU()

# Aqui estou modificando os pesos e viés que seriam gerados 
# de maneira aleatória pelos que eu já havia pré-definido anteriormente. 
perceptron.weight = nn.Parameter(torch.Tensor([[w1,w2]]))
perceptron.bias = nn.Parameter(torch.Tensor([b]))


makers = ['^', 'v', '>', '<']
colors = ['g', 'b', 'r', 'y']

plt.figure(figsize=(10,8))
plotmodel(w1,w2,b)

for k, idx in enumerate([17,21,43,66]):
  
  # Aqui estarei armazenando as coordenadas de
  # cada elemento, do 17 ao 66.
  x = torch.Tensor(X[idx])

  # O que estou nomeando como retorno são os dados
  # que passaram pelo perceptron, que são retornados
  # cada quais com seu respectivo valor e, assim, 
  # passando-os para a função de ativação. 
  retorno = perceptron(x)
  funcao_ativada = ativando(retorno)

  ativacao_limiar = 0 if retorno.data < 0 else 1

  label = "retorno: {:5.2f}".format(retorno.data.numpy()[0]) + "; limiar {:4.2f}".format(ativacao_limiar) + "; função relu: {:5.2f}".format(funcao_ativada.data.numpy()[0])
  plt.plot(x[0], x[1], marker = markers[k], color = colors[k], markersize=10, label = label)

plt.legend()
plt.show

# Aqui o comportamento da relu é diferente. Ela ignora valores negativos, 
# atribuindo a eles um retorno de zero, porém para os positivos mantém seus 
# respectivos valores na sua saida, não se restringindo a apenas ao 1 como saida. 
```
